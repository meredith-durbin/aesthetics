```python
# python 2.7 kernel, matplotlib 1.5.3

%matplotlib inline
%config inlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import numpy as np
from aesthetics import aestheticize
```

```python
def testplot():
    # set up data
    x = np.linspace(0,10,51)
    y = x**2
    y2 = 100 - y
    imgdata = np.multiply.outer(x,x)
    g1 = np.random.normal(size=1000)
    g2 = np.random.normal(size=1000)
    
    # 2x3 subplots
    fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2,3,figsize=(9,6),dpi=100)
    
    # line plot, scatterplot
    for i in range(1,11):
        ax1.plot(x,y2*i,label=i)
        ax1.scatter(x[::4],y[::4]*i)
    ax1.legend(ncol=2)
    ax1.set_title('plot, scatter')

    # image plot
    imgdata = np.multiply.outer(x,x)
    ax2.imshow(imgdata,vmin=0,vmax=70)
    ax2.set_title('imshow')
    
    # filled contour
    ax3.contourf(imgdata)
    ax3.set_title('contourf')

    # 2D histogram
    g1 = np.random.normal(size=1000)
    g2 = np.random.normal(size=1000)
    ax4.hist2d(g1,g2,bins=20)
    ax4.set_title('hist2d')

    # 1D histogram
    ax5.hist(imgdata.ravel())
    ax5.hist(imgdata.max() - imgdata.ravel(), histtype='step')
    ax5.set_yscale('log')
    ax5.set_title('hist')
    
    # fill between with two y values
    for i in range(1,9):
        scale = 1./i
        offset = 50-50./i
        ax6.fill_between(x,y*scale+offset,y2*scale+offset)
    ax6.set_title('fill\_between')
```

```python
testplot()
```

```python
aestheticize(colorscheme='brewer',verbose=True)
testplot()
```

```python
aestheticize(colorscheme='vaporwave',verbose=True)
testplot()
```

```python
aestheticize(colorscheme='bw',verbose=True)
testplot()
```
